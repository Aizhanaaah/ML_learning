```python
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import scatter_matrix
```

```python
cols = ['Top10perc', 'Apps', 'Enroll']
scatter_matrix(college[cols], figsize=(8, 8), alpha=0.7, diagonal='hist')
plt.show()
```

```python
college.boxplot(column="Outstate", by="Private", figsize=(8, 8))
plt.title("Outstate vs Private")
plt.xlabel("Private")
plt.ylabel("Outstate")
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
[0,50,100],
labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite', figsize=(8, 8))
plt.title('Outstate vs Elite')
plt.xlabel('Elite')
plt.ylabel('Outstate')
plt.show()
```

```python
college.plot.hist(bins=20, alpha=0.7, color="skyblue")
```

```python
cols = ["Apps", "Accept", "Enroll", "Outstate"]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes = axes.ravel()

for i, col in enumerate(cols):
    college[col].plot.hist(
        bins=10*(i+1),    
        alpha=0.7,
        color="skyblue",
        edgecolor="black",
        ax=axes[i]
    )
    axes[i].set_title(f"{col} (bins={10*(i+1)})")

plt.tight_layout()
plt.show()
```
